# <center>Improving ICA with optimisation</center>

In this notebook we implement a homemade ICA based on J. Shlens tutorial [1] with a fourth-order blind identification (FOBI) method as depicted in [2] to start solving the statistical independence. Then we use various kind of constraint optimizations to improve the result.

# Context

Let's build different type of situations where the sources that generate the data measured are known. We will first tackle situations with 2 sources and 2 sensors since the data is very easy to visualize in such situation. Samples of the 2 sources $S$ will be carried out for various type of distribution:
- sharply peaked sources
- unimodal and bimodal sources
- gaussian sources

The corresponding data measured $X$ stems from a linear mixture of sources $S$ perturbed by some noise $\epsilon$:

\begin{equation}
X=AS+\epsilon
\end{equation}

With $X$ and $S$ being arrays of shape:
- X: (d,n) where *d* is the dimension of the measures : we have 2 sensors.
- S: (l,n) where *l* is the number of sources.
- *n* is the number of samples carried out in the measures.

# Introduction with a first example

```python
# Before anything, import relevant modules
%load_ext autoreload
%autoreload 2

import autograd.numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from utils import plot_2D_distrib, mutual_information, analyze
from core import optim_ICA

rc('text', usetex=True)
```

Let's generate some sources and measures:

```python
from benchmark import get_gaussian_noise, get_sharp

n = 1000
d = 2
noise_ratio = 0/100
A = np.array([[   2,  1],
              [ 0.5, -4]])
k = 100

noise = get_gaussian_noise(n=n, d=d, cov=noise_ratio*np.ones(2))
S = get_sharp(n=n, d=d, k=0.2, theta=1)

X = A.dot(S) + noise

print('Covariance estimation of used sources:\n', S.dot(S.T))
```

> **NOTE**. As we can see, even if the sources $s_1$ and $s_2$ have been generated independantly, their covariance estimation is quite far from the identity matrix...

```python
ax = plot_2D_distrib(X, title='measured', k=k)
ax = plot_2D_distrib(S, title='sources', k=k)
ax = plot_2D_distrib(noise, title='noise', k=k)
```

```python
from heuristics import FOBI, V_manifold_optim
```

We import two methods used during ICA computation.
- The first one is **FOBI**, an algebric method based on 4th order of whitened data. This will provid a first estimation of $A$ and $S$ that will be stored in results.
- Then **V_manifold_optim** will take over, returning a new estimation, more accurate, of $A$ and $S$, added to the results returned.

The methods are passed in the argument *schedule* and will be processed in the FIFO order. Their possible *kwargs* are passed in the field *schedule_kwargs*.

```python
res, params = optim_ICA(X,
                        schedule=[FOBI, V_manifold_optim],
                        schedule_kwargs=[dict(), {'grad_eps': 1e-1, 'k': 100, 'reg': False}],
                        verbose=True)
```

We can compare this last plot of estimated sources to the one actually used to generate the data:

```python
ax = plot_2D_distrib(S, title='sources', k=k)
```

And print the results obtained through the various methods specified in *schedule*:

```python
res
```

And compare the estimated linear transformation $\hat{A}$ with the one really used $A$:

```python
print('A_hat = \n',res[-1]['A'],'\n\nA = \n',A)
```

As we can see, the estimated linear transformation $\hat{A}$ might differ from the one actually used in the benchmark: $A$. Indeed, the problem is not well posed, the estimated sources are always up to a permutation. Moreover each source is up to a scale factor. By imposing a whitened source estimation $\hat{S}$ in the sense $$\hat{S}\hat{S}^T = I_l$$, the scale factors are passed on the estimated linear transformation $\hat{A}$. As shown below:
- we can build and analyze $\hat{A}_{matched}$, a good estimation of $A$ and $\hat{S}_{matched}$ a good estimation of $S$ simply by looking at the matrix $\text{cov}(S, \hat{S})$ to permute and rescale the sources.
- the couple $\hat{A}, \hat{S}$ fits well the measured data in the sense $X\simeq\hat{A}\hat{S}$
- the statistical independance between sources in $S$ is well respected

```python
from benchmark import match_with_sources

result = res[-1]
solution = {'A': A, 'S': S}
a_hat_matched, s_hat_matched, _, _ = match_with_sources(result=result, solution=solution, verbose=True)

print('A_hat_matched = \n',a_hat_matched,'\n\nA = \n',A)
```

**<center>Now $\hat{A}_{matched}$ and $A$ matched pretty well!</center>**  
<center>And so are doing $\hat{S}_{matched}$ and $S$:</center>

```python
ax = plot_2D_distrib(s_hat_matched, 'estimated matched sources', k=k)
ax = plot_2D_distrib(S, title='true sources', k=k)
```

<center>And of course both $\hat{A}\hat{S}$ and $\hat{A}_{matched}\hat{S}_{matched}$ are perfectly matching $X$ since no noise assumption was made:

```python
print('\n||A_hat.S_hat-X||/||X||=', 
      (np.average(((res[-1]['A'].dot(res[-1]['S'])-X))**2)/np.average(X**2))**.5, '%')
print('\n||A_hat_macthed.S_hat_macthed-X||/||X||=', 
      (np.average(((a_hat_matched.dot(s_hat_matched)-X))**2)/np.average(X**2))**.5, '%') 
```

>**NOTE**: Why is the match not 100%?!

```python
# Unimodal and bimodal sources
s1 = np.random.normal(0,1,n)
s2 = np.random.gamma(15,0.1,n)*(2*np.random.randint(2, size=n)-1)

# substracting sources mean
s1 -= np.average(s1)
s2 -= np.average(s2)

# normalizing variance of sources
s1 /= (s1.dot(s1.T))**.5   
s2 /= (s2.dot(s2.T))**.5

S = np.array([s1,s2])

A = np.array([[   2, 1],
              [ 0.5, 3]])

X = A.dot(S)

res, params = optim_ICA(X)
```

```python
print('A_hat = \n',res[-1]['A'],'\n\nA = \n',A)
```

> **NOTE**. As shown bellow, gaussians sources are known to be a limit of ICA, they can not be unmixed ...

```python
# Gaussian sources
s1 = np.random.normal(10,1,n)
s2 = np.random.normal(-2,5,n)

# substracting sources mean
s1 -= np.average(s1)
s2 -= np.average(s2)

# normalizing variance of sources
s1 /= (s1.dot(s1.T))**.5   
s2 /= (s2.dot(s2.T))**.5

S = np.array([s1,s2])

A = np.array([[   2, 1],
              [ 0.5, 3]])

X = A.dot(S)

res, params = optim_ICA(X)
```

```python
print('A_hat = \n',res[-1]['A'],'\n\nA = \n',A)
```
